```python
import pandas as pd

df = pd.read_csv("products_catalog.csv")

print("==== 5 Baris Pertama ====\n", df.head())
print("\n==== Info Dataset ====\n", df.info())
print("\n==== Statistik Deskriptif ====\n", df.describe(include="all"))
print("\n==== Nama Kolom ====\n", df.columns)
```

```python
#Cek Missing Value
print("==== Missing Values ===\n=", df.isnull().sum())
print("\n")

# Tangani missing values
df['ProductBrand'] = df['ProductBrand'].fillna("Unknown")
df['Gender'] = df['Gender'].fillna("Unisex")
df['PrimaryColor'] = df['PrimaryColor'].fillna("Unknown")
print("Missing Values setelah imputasi:")
print(df.isnull().sum())
print("\n")

# Hapus duplikat berdasarkan ProductID
before = len(df)
df = df.drop_duplicates(subset='ProductID')
after = len(df)
print(f"Duplikat dihapus: {before - after}, Sisa data: {after}")
print("\n")

# Drop baris jika harga kosong
before = len(df)
df = df.dropna(subset=['Price (INR)'])
after = len(df)
print(f"Baris dihapus karena harga kosong: {before - after}, Sisa data: {after}")
print("\n")

# Pastikan tipe data benar
df['Price (INR)'] = pd.to_numeric(df['Price (INR)'], errors='coerce')
df['NumImages'] = pd.to_numeric(df['NumImages'], errors='coerce')
print("Tipe data setelah konversi:")
print(df.dtypes)
print("\n")

# Final check
print("==== Missing Values (final) ====")
print(df.isnull().sum())
print("\n")

print("==== Preview Data ====")
print(df.head())
```
